## Item b

```r
fit1<-VAR(AFP2, lag.max=10 , type = "const")
fit1$p
```

```r
VARselect(AFP2, lag.max=10,type='const')
```

Según los metodos conocidos el parametro p debería ser 9, 1,1 u 8. Algo muy extraño, pues los numeros no son cercanos. Por otro lado tenemos que el 9 y el 8 responden fluctuaciones pequeñas dentro de sus metodos, veamos que si tomaramos un lag maximo igual a 5 obtendiramos que el modelo ideal es con 1.

```r
fit2<-VAR(AFP2, lag.max=5 , type = "const")
fit2$p
```

```r
VARselect(AFP2, lag.max=5,type='const')
```

## Item E

```r
nts<-window(ts,start=start(ts),end=c(2019,5))
fit3.1<-VAR(nts, lag.max=1 , type = "const")
fit3.1$p
fit3.9<-VAR(nts, lag.max=10 , type = "const")
fit3.9$p
```

```r
data_oculta<-window(ts,start=c(2019,6),end=end(ts))
do<-data.frame(data_oculta)
do
```

```r
do$Cuprum[3]**2
```

```r
predict(fit3.1,n.ahead=4,ci=0.95)
```

```r
temp=c()

for (j in c(1,2,3,4)){
    temp=c(temp,(predict(fit3.1,n.ahead=4,ci=0.95)$fcst$Habitat.A[j]-do$Habitat.A[j])**2)}
error_cuad_med_Habitat.a<-sum(temp)/4
```

Mientras para el modelo VAR(9) tenemos problemas para garantizar esto con el estadistico Skewness para una significancia menor al 8%, lo que da más razones para considerar que el modelo VAR(1) es el ideal para este contexto
