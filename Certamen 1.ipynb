```r
I=diag(9)
v_sigma<-c(0.26,0.03,0,0.03,0.09,0,0,0,0.81) #Vectorización de la matriz de covarianzas
inv<-solve(I-AA)
inv
```

```r
gamma_0=matrix(inv%*%v_sigma,ncol=3)
gamma_0
```
