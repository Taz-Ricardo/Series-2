## a) realizar ajuste GARCH

```r
ventas<-sales
ventas2=ventas^2
```

```r
ts.plot(ventas)
par(mfrow=c(1,2))
acf(ventas)
pacf(ventas)
```

Vemos que el candidato a modelo según la función de autocorrelación parcial sugiere un modelo GARCH(1,0)

```r
ts.plot(ventas2)
par(mfrow=c(1,2))
acf(ventas2)
pacf(ventas2)
```

```r
model1=garchFit(~ garch(1,0), data = ventas, trace = FALSE)
summary(model1)
```

```r
model2=garchFit(~ garch(2,0), data = ventas, trace = FALSE)
summary(model2)
```

```r
model3=garchFit(~ garch(1,1), data = ventas, trace = FALSE)
summary(model3)
```

## Conclusión:
Vemos que el mejor modelo garch es el GARCH(1,0), pero no es un buen modelo, de hecho vemos que los residuos no distribuyen normal, para las significancias típicas para los test JB y Shapiro Wilk, mientras el efecto ARCH no pareciera ser influyente, es decir, aún con "el mejor" el ajuste es malo.
