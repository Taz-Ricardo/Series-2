## Cuprum

```r
count_maV1 = ts(na.omit(AFP$V1), frequency=2)
decompV1 = stl(count_maV1, s.window="periodic")
deseasonalV1 <- seasadj(decompV1)
plot(decompV1)
```

```r
modeloarimaV1<-auto.arima(deseasonalV1, seasonal=FALSE)
modeloarimaV1
```

```r
tsdisplay(residuals(modeloarimaV1), lag.max=10, main='(0,0,1) Model Residuals V1')
```

El modelo pareciera ser satisfactorio

## Comentario
Hice un for que me generara todas las combinaciones de AFPs posibles, no pude cambiar los titulos de los graficos, de igual manera, construí una matriz que guarda todos los valores, la matriz tiene el coeficiente de codisperción de V1 y V2, luego V1 y V3 y así, la matriz esta abajo de todos los graficos

```r
library(stringr) #Libreria para poder concadenar strings, finalmente no se utilizo, porque no pude cambiar el titulo de los graficos
```

```r
coords <- cbind(1:180, rep(1,180))
Z=c()
```

```r
temp1=0
for (i in AFP){
    temp1=temp1+1
    temp2=1
    for (j in AFP){
    if (temp1<temp2){
        z <- codisp(i, j,coords)
        z
        plot(z)
        Z=c(Z,z$coef)
        }
#        plot(z, main=str_c('Codisperción entre', V[temp1], 'y', V[temp2], sep = ", "))}
# Código para cambiar nombre de los graficos, que no funciona al parecer, el paquete utilizado no permite editar el nombrre de los graficos
        temp2=temp2+1}
}
```

```r
length(z$coef)
```

```r
Zm=matrix(Z, nrow=13)
```

```r
Zm #Matriz que guarda los comovimientos.
```

```r
range(Zm)
```

### Observaciones:
Pareciera que existe una tendencia a la baja de la seire y estacionalidad de 12 meses, junto con un comportamiento que podría ser debido a las estaciones del año, en particular el invierno sería un periodo de interes, por lo que sería lógico un regimen de humbrales (modelos TAR).

```r
flu
```

## Enfoque Heuristico
Según el criterio de la raiz al numerro de la muestras, debe ser menor o igual a 11, P=11 ,considerando el grafico y que tengo una relación entre años quisiera tomar $p_1=9$, $p_2=3$ esperando una relación entre las estaciones del año y ajustes para el modelo, d=12 por ser valores anuales y me fijo $R_1=(-\infty,0.3)$ y $R_2=(0.3,\infty)$

```r
library(TSA) #Paquete que tiene implementado los modelos TAR
```

```r
tar(flu, 9, 3, 12, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.3, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = TRUE)
```

```r
tar(flu, 8, 3, 12, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.3, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = TRUE)
```

### Conclusiones Heuristica de primer acercamiento

Podemos observar que el comando utilizado realiza un ajuste de cuantas regresoras tener, pues para $p_1=9$ entrega el mismo resultado que para $p_1=8$, los cuales tienen un AIC bajo, además que el estadistico F da claras muestras que los modelos son utiles, pues su valor p es practicamente 0 para ambos casos. Además las pruebas t parciales en todos los casos con una significancia del 5% rechazo $H_0: \beta _i=0$. Este sería un primer acercamiento, al "ojo".

### Metodología Heuristica vista en clases

Pero ahora si los valores de $p_i$ son cotas de los valores, buscando minimizar el AIC, utilicemos la cota dada por Tong, que es $P=11$

```r
tar(flu, 11, 11, 12, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.3, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = TRUE)
```

Vemos que la significancia de los test F no ha cambiado, sigue siendo 0, tenemos un mejor AIC, pero ahora las pruebas t parciales, no son tan buenas como antes en todos los casos, de hecho hay algunos valores que se rechazarían para cualquier significancia típica.

Siguiendo la cadena, ya tenemos los valores de los $p_j$ por lo que buscaremos el valor para r y para esto recorreremos el intervalo $(0.25, 0.35)$ de saltos de 0.01 para ver el mejor r dados los p_j.


## Estimación del valor r.


```r
AICR=c()
mallado_R=c(0.25,0.26,0.27,0.28,0.29,0.30,0.31,0.32,0.33,0.34,0.35)
for (i in mallado_R){
AICR=c(AICR,tar(flu, 11, 8, 12, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=i, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = FALSE)$AIC)}
df_AICR<-data.frame(mallado_R,AICR)
df_AICR
min(df_AICR$AICR)
#No recuerdo como obtener la otra componente del data frame, cuando hago el minimo, si usted me podría explicar en un futuro, le agradecería mucho.
```

Con lo que tenemos que el valor $r=0.28$ es el valor con menor AIC.

## Estimación del valor d.


```r
AICD=c()
mallado_D=c(1:20)
for (i in mallado_D){
AICD=c(AICD,tar(flu, 11, 8, i, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.28, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = FALSE)$AIC)}
df_AICD<-data.frame(mallado_D,AICD)
df_AICD
```

```r
AICDnd=c()
n=132
for (i in mallado_D){
    nd=max(11,i)
    temp=AICD[i]/(n-nd)
    AICDnd=c(AICDnd,temp)}
df_AICDnd<-data.frame(df_AICD,AICDnd)
min(df_AICDnd$AICDnd)
df_AICDnd
#No recuerdo como obtener la otra componente del data frame, cuando hago el minimo, si usted me podría explicar en un futuro, le agradecería mucho.
```

Con lo que tenemos que nuestra estimación de d es $\hat{d} =12$.
