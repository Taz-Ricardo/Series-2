## a) realizar ajuste GARCH

```r
ventas<-sales
ventas2=ventas^2
```

```r
ts.plot(ventas)
par(mfrow=c(1,2))
acf(ventas)
pacf(ventas)
```

```r
ts.plot(ventas2)
par(mfrow=c(1,2))
acf(ventas2)
pacf(ventas2)
```

```r
model1=garchFit(~ garch(1,0), data = ventas, trace = FALSE)
summary(model1)
```

```r
model2=garchFit(~ garch(2,0), data = ventas, trace = FALSE)
summary(model2)
```

```r
model3=garchFit(~ garch(1,1), data = ventas, trace = FALSE)
summary(model3)
```

### c) Comprobar computacionalmente que el proceso al cuadrado sigue un modelo autoregresivo.

```r
ts.plot(ventas2)
par(mfrow=c(1,2))
acf(ventas2)
pacf(ventas2)
```

Podemos ver que un proceso ar(1) tendría sentido, probemos utilizando la función autoarima

```r
ar1<-arima(ventas2,c(1,0,0))
ar1
```

Podemos ver que el proceso al cuadrado cumple con ser un procesor del tipo ar(1)
